Q1.What is a random variable in probability theory?

Q7.What are the key properties of a Bernoulli distribution?

Q8.What is the binomial distribution, and how is it used in probability?

Q10.What is a continuous uniform distribution?

Q11.What are the characteristics of a normal distribution?

.Normal distributions have key characteristics that are easy to spot in graphs: The mean, median and mode are exactly the same. The distribution is symmetric about the mean—half the values fall below the mean and half above the mean. The distribution can be described by two values: the mean and the standard deviation.

Q12.What is the standard normal distribution, and why is it important?

Q13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?

.The Central Limit Theorem (CLT) is a fundamental statistical concept stating that the distribution of sample means will approximate a normal distribution, regardless of the original population distribution, as the sample size increases. It's critical because it allows for statistical inferences and hypothesis testing, even when the original data isn't normally distributed.



Q14.How does the Central Limit Theorem relate to the normal distribution?

.The central limit theorem says that the sampling distribution of the mean will always be normally distributed, as long as the sample size is large enough. Regardless of whether the population has a normal, Poisson, binomial, or any other distribution, the sampling distribution of the mean will be normal.

.The formula for calculating a z-score is z = (x-μ)/σ, where x is the raw score, μ is the population mean, and σ is the population standard deviation. As the formula shows, the z-score is simply the raw score minus the population mean, divided by the population standard deviation.


Q17. What are point estimates and interval estimates in statistics?

Q18.What is the significance of confidence intervals in statistical analysis?

Q19. What is the relationship between a Z-score and a confidence interval?

Q20.How are Z-scores used to compare different distributions?

Formula:

Q3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    """
    Calculate the Probability Distribution Function (PDF) of a Bernoulli distribution.

    Parameters:
    - x: The outcome (either 0 or 1)
    - p: The probability of success (0 <= p <= 1)

    Returns:
    - Probability of the outcome x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1 (inclusive)")


    if x == 1:
        return p
    else:
        return 1 - p

p = 0.7
x = 1
print(f"P(X = {x}) = {bernoulli_pdf(x, p)}")

x = 0
print(f"P(X = {x}) = {bernoulli_pdf(x, p)}")
```

Q9.Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate the Z-scores for a given dataset.

    Parameters:
    - data: A list or array of numerical data.

    Returns:
    - z_scores: An array of Z-scores for the data.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores


data = np.random.normal(0, 1, 1000)

z_scores = calculate_z_scores(data)


plt.plot(z_scores, label='Z-scores')
plt.axhline(0, color='red', linestyle='--', label='Mean (Z = 0)')
plt.xlabel('Index')
plt.ylabel('Z-score')
plt.title('Z-scores of the Dataset')
plt.legend()
plt.grid(True)
plt.show()
```

Q18.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    - data_point: The data point for which the Z-score is calculated.
    - mean: The mean of the data.
    - std_dev: The standard deviation of the data.

    Returns:
    - z_score: The Z-score for the data point.
    """
    z_score = (data_point - mean) / std_dev
    return z_score

def plot_standard_normal_distribution():
    """
    Plot the standard normal distribution (mean=0, std_dev=1).
    """

    x_values = np.linspace(-4, 4, 1000)
    y_values = norm.pdf(x_values, loc=0, scale=1)

    plt.plot(x_values, y_values, label='Standard Normal Distribution', color='blue')
    plt.fill_between(x_values, y_values, color='lightblue', alpha=0.5)
    plt.axvline(x=0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.title('Standard Normal Distribution (mean=0, std=1)')
    plt.grid(True)
    plt.legend()
    plt.show()


data_point = 2
mean = 0
std_dev = 1

z_score = calculate_z_score(data_point, mean, std_dev)

cumulative_prob = norm.cdf(z_score, loc=mean, scale=std_dev)


print(f"Data point: {data_point}")
print(f"Z-score: {z_score:.2f}")
print(f"Cumulative probability (P(Z <= {z_score:.2f})): {cumulative_prob:.4f}")


plot_standard_normal_distribution()
```

Q19. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, sample_size, alpha=0.05):
    """
    Perform Z-test for hypothesis testing on sample data.

    Parameters:
    - sample_data: The sample data
    - population_mean: The mean of the population (under H0)
    - population_std: The population standard deviation
    - sample_size: The number of observations in the sample
    - alpha: Significance level (default is 0.05)

    Returns:
    - z_stat: The calculated Z-statistic
    - p_value: The p-value for the test
    - result: Result of the hypothesis test (reject or fail to reject H0)
    """

    sample_mean = np.mean(sample_data)


    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))


    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    critical_value = norm.ppf(1 - alpha / 2)

    if abs(z_stat) > critical_value:
        result = "Reject the null hypothesis"
    else:
        result = "Fail to reject the null hypothesis"

    return z_stat, p_value, result


sample_data = [67, 69, 64, 70, 66, 68, 64, 65, 67, 71, 62, 69, 65, 63, 66, 68, 67, 64, 65, 67, 68, 70, 69, 72, 66, 68, 64, 69, 67, 65]
population_mean = 65
population_std = 5
sample_size = len(sample_data)


z_stat, p_value, result = z_test(sample_data, population_mean, population_std, sample_size)


print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Result: {result}")
```

Q20. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence_level=0.95):

    sample_std = np.std(data, ddof=1)
    sample_size = len(data)


    standard_error = sample_std / np.sqrt(sample_size)


    z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)


    margin_of_error = z_value * standard_error


    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean, margin_of_error


data = [67, 69, 64, 70, 66, 68, 64, 65, 67, 71, 62, 69, 65, 63, 66, 68, 67, 64, 65, 67, 68, 70, 69, 72, 66, 68, 64, 69, 67, 65]

lower, upper, mean, margin = calculate_confidence_interval(data)

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print(f"Margin of Error: {margin:.2f}")
```

Q21.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.


```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence_level=0.95):

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)

    standard_error = sample_std / np.sqrt(sample_size)

    z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)


    margin_of_error = z_value * standard_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean, margin_of_error


np.random.seed(42)
population_mean = 100
population_std = 15
sample_size = 50

sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

lower, upper, mean, margin = calculate_confidence_interval(sample_data)


print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print(f"Margin of Error: {margin:.2f}")
```

Q28.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    """
    Function to calculate skewness and kurtosis of a dataset.
    """

    data_skewness = skew(data)


    data_kurtosis = kurtosis(data)

    return data_skewness, data_kurtosis


np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

skewness_value, kurtosis_value = calculate_skewness_kurtosis(data)


print(f"Skewness of the dataset: {skewness_value:.4f}")
print(f"Kurtosis (Excess) of the dataset: {kurtosis_value:.4f}")
```
